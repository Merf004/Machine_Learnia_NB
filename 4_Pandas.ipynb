## Pandas et series temporelles

### Analyse des donnÃ©es du bitcoin

```python
bitcoin = pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)
bitcoin
# index_col = 'Date': Pour mettre en index la colonne des dates
# parse_dates = True: Pour changer les tranformer en date(etant au prÃ©alable en str) 
```

```python
# Evolution du bitcoin en 2019
bitcoin.loc['2019']['Close'].plot()
```

```python
# Evolution du bitcoin en sept 2015
bitcoin.loc['2015-09']['Close'].plot()
```

### Fonction resample()

Elle permet de regrouper les donnÃ©es selon une frequence temporelle

```python
# Regroupement des donnÃ©es bitcoin en 2019 en moyenne par mois
bitcoin.loc['2019']['Close'].resample('M').mean().plot()
```

```python
# Regroupement des donnÃ©es bitcoin en 2019 en moyenne tous les trimestres
bitcoin.loc['2019']['Close'].resample('3M').mean().plot()
```

```python
# Regroupement des donnÃ©es bitcoin en 2017 en moyenne tous les 5 jours
bitcoin.loc['2019']['Close'].resample('5D').mean().plot()
```

### Fonction aggregate()

Elle fait la meme chose que resample mais permet d'avoir plusieurs mesures statistiques

```python
# Regroupement des donnÃ©es bitcoin en 2019 par mois
bitcoin.loc['2019']['Close'].resample('M').agg(['mean', 'std', 'min', 'max'])
```

```python
# Regroupement des donnÃ©es bitcoin en 2019 par mois
bitcoin.loc['2019']['Close'].resample('M').agg(['mean', 'std', 'min', 'max']).plot()
```

### Moyenne mobile

Faire la moyenne sur une fenetre de valeur, en avance d'une unitÃ© dans le graphe. Oun utilise une fonction rolling

```python
# Creation de la fonction rolling avec une fentre de 7
bitcoin.loc["2019"]['Close'].rolling(window = 7).mean().plot()
# On peut l'utiliser avec d'autre statistique(ecart-type, min, max)
```

```python
plt.figure(figsize = (12,8))
bitcoin.loc['2019']['Close'].plot(label = 'Close')
bitcoin.loc["2019"]['Close'].rolling(window = 7).mean().plot(label = 'Moyenne mobile', ls=':')
plt.legend()
```

```python
# Parametre center pour centrer le resulata de la moyenne mobile, et non le mettre Ã  la fin

plt.figure(figsize = (12,8))
bitcoin.loc['2019-09']['Close'].plot(label = 'Close')
bitcoin.loc["2019-9"]['Close'].rolling(window = 7, center = True).mean().plot(label = 'Moyenne mobile: center = True', ls=':')
bitcoin.loc["2019-9"]['Close'].rolling(window = 7).mean().plot(label = 'Moyenne mobile: center = False', ls=':')
plt.legend()
```

### Moyenne mobile exponentielle

Elle suit la formule: MME = alpha(xt + (1 - alpha)xt-1 + (1-alpha)**2xt-2 + (1 - alpha)**3xt-3) + ...)

```python
plt.figure(figsize = (12,8))
bitcoin.loc['2019-09']['Close'].plot(label = 'Close')
bitcoin.loc["2019-9"]['Close'].rolling(window = 7, center = True).mean().plot(label = 'Moyenne mobile: center = True', ls=':')
bitcoin.loc["2019-9"]['Close'].rolling(window = 7).mean().plot(label = 'Moyenne mobile: center = False', ls=':')
bitcoin.loc["2019-9"]['Close'].ewm(alpha = 0.6).mean().plot(label = 'Moyenne mobile exponentielle', ls=':')
plt.legend()
```

### Assembler des datasets

```python
ethereum = pd.read_csv('ETH-EUR.csv', index_col='Date', parse_dates=True)
ethereum
```

```python
ethereum.loc['2019']['Close'].plot()
```

```python
# Assemblage avec inner
btc_eth = pd.merge(bitcoin, ethereum, on='Date', how='inner', suffixes=('_btc', '_eth'))
btc_eth

# Les donnÃ©es bitcoin commencent en 2011 et les donnÃ©es ethereum en 2015. Avec inner, on supprimme toutes les donnÃ©es qui ne sont pas en
# commun. Donc toutes les lignes avec date de bitcoin < 2015
```

```python
# Assemblage avec outer
pd.merge(bitcoin, ethereum, on='Date', how='outer', suffixes=('_btc', '_eth'))

# Avec outer, on assemble toutes les donnÃ©es meme ceux qui ne sont pas en commun et on met des NaN
```

```python
btc_eth[['Close_btc', 'Close_eth']].plot(subplots = True, figsize = (12,8))
```

```python
# Correlation entre le bitcoin et l'ethereum
btc_eth[['Close_btc', 'Close_eth']].corr()
```
