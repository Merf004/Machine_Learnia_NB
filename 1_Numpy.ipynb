```python
# Fonction pour l'exercice

def init(m, n):
    A = np.random.rand(m, n)
    B = np.ones((m, 1))
    C = np.concatenate((A,B), axis = 1)
    return C
```

```python
E = init(5,6)
```

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray = True)
plt.imshow(face, cmap = plt.cm.gray)
plt.show()
```

```python
face.shape
```

```python
h = face.shape[0]
l = face.shape[1]

zoom = face[h//4:-h//4, l//4:-l//4]
```

```python
plt.imshow(zoom, cmap = plt.cm.gray)
plt.show()
```

```python
zoom[zoom < 75] = 0
zoom[zoom > 200] = 255

plt.imshow(zoom, cmap = plt.cm.gray)
plt.show()
```
